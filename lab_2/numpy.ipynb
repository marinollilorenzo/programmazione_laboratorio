- numPy per il calcolo numerico in Python
- progettata per il calcolo statistico e utilizzato per l'analisi dei dati.
- gestisce in modo super efficiente array e prodotti matriciali di grandi dimensioni e multidimensionali
- Alcuni casi d'uso specifici di NumPy per l'analisi dei dati:

> 4. Generazione di numeri casuali.

```python
import numpy as np
```

```python
import random

lst_numbers = [random.randint(50000, 150000) for _ in range(1000000)]
```

```python
import statistics
```

```python
%%timeit

statistics.mean(lst_numbers)
```

```python
%%timeit

np.mean(lst_numbers)
```
